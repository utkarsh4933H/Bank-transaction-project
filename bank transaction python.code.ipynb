```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('bank_transactions.csv')
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['CustLocation'] = df['CustLocation'].fillna('Not available')
```

```python
df['CustomerDOB'] = df['CustomerDOB'].fillna(1/1/2008)
```

```python
df['CustGender'] = df['CustGender'].fillna(df['CustGender'].mode()[0])
```

```python
df['CustAccountBalance']=df['CustAccountBalance'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'transactionamount_(inr)':'transaction_amount'})
```

```python
df.columns
```

```python
#convert the customerdob to datetime format  1/1/2005 se  2005-01-10
df['customerdob']=pd.to_datetime(df['customerdob'],format='%d/%m/%y', errors='coerce')
```

```python
#create a age columns extracting the age from the customerdob column  EXAMPLE 9/1/2005 se  2005-01-10 se 2024-06-01 se age 19
today = datetime.today()
df['age'] = (today - df['customerdob']).dt.days // 365
```

```python
#create a age columns 
labels = ['young','middle_aged','old']
df['age_groups'] = pd.qcut(df['age'], q=3, labels=labels)
```

```python
df[['age_groups', 'age']].head()
```

```python
# time feature 
df['transactiondate'] = pd.to_datetime(df['transactiondate'],format='%d/%m/%y', errors='coerce')
df['month'] = df['transactiondate'].dt.month
df['day']= df['transactiondate'].dt.day
```
